```python
from rectools.models.implicit_als import ImplicitALSWrapperModel
from implicit.als import AlternatingLeastSquares
from rectools.dataset import Dataset
from rectools import Columns
from rectools.models import load_model
import pandas as pd
```

```python
df = pd.read_csv(
        '/Users/nikitaartamonov/jupyter-projects/mlops/project/MovieRecs/data/raw/ratings.dat', 
        sep='::', 
        engine='python', 
        header=None, 
        names=[Columns.User, Columns.Item, Columns.Weight, Columns.Datetime],
        encoding='latin-1'
        )
```

```python
df[Columns.Weight] = 1.0
```

```python
train_set = Dataset.construct(df)
```

```python
model = ImplicitALSWrapperModel(AlternatingLeastSquares(factors=200, alpha=40, num_threads=3),
                                                        recommend_n_threads=3, 
                                                        verbose=1)
```

```python
model.fit(train_set)
```

```python
model.save('/Users/nikitaartamonov/jupyter-projects/mlops/project/MovieRecs/models/ials.pkl')
```
